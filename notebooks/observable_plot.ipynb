# Embed the same chart using our data

```python
# First get our data
import seaborn as sns

iris = sns.load_dataset('iris')
```

```python
# Define the configurable parameters, which includes our dataset
params = {
    "csv": iris.to_csv(),
    "x": "sepal_length",
    "y": "sepal_width",
    "stroke": "species",
    "fill": "species"
}
```

```python
# Without specifying cells explicitly, you'll load the entire notebook
#embed("@pbogden/observable-plot-jupyter", inputs=params)
```

```python
# Load the chart with our dataset
embed("@pbogden/observable-plot-jupyter", cells=["chart"], inputs=params)
```
